```python
from databricks.sdk import WorkspaceClient
from databricks.sdk.service.catalog import AzureManagedIdentity

w = WorkspaceClient()

w.storage_credentials.create(
    name="my_cred",
    azure_managed_identity=AzureManagedIdentity(
        access_connector_id="/subscriptions/3106f1bf-7bd3-4842-a714-5ba9a6ba8da1/resourceGroups/EndtoEndProject/providers/Microsoft.Databricks/accessConnectors/accessconnectorproject"
    ),
)
```

```python
%sql
CREATE EXTERNAL LOCATION source_ext
URL 'abfss://source@storageaccountfull.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL my_cred);
```

```python
%sql
CREATE EXTERNAL LOCATION bronze_ext
URL 'abfss://bronze@storageaccountfull.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL my_cred);

CREATE EXTERNAL LOCATION silver_ext
URL 'abfss://silver@storageaccountfull.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL my_cred);

CREATE EXTERNAL LOCATION gold_ext
URL 'abfss://gold@storageaccountfull.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL my_cred);
```
